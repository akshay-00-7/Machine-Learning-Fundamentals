```python
df.columns
```

```python
x =df[["Gender", "Age", "Height", "Weight", "Duration",
       "Heart_Rate", "Body_Temp"]]
x
```

```python
y = df[	"Calories"  ]
y
```

```python
gender_dummies = pd.get_dummies(x["Gender"], drop_first=True).astype(int)
```

```python
x = x.drop("Gender", axis=1)
x = pd.concat([x, gender_dummies], axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split (x,y ,test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression 

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

y_train_pred = model.predict(x_train)

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

scatter = plt.scatter(
    x_train["Duration"],
    y_train,
    c=x_train["Heart_Rate"],      # color = heart rate
    s=x_train["Body_Temp"] * 5,   # size = body temp
    alpha=0.6
)

plt.xlabel("Duration")
plt.ylabel("Calories")
plt.title("Training Data: Duration + Heart Rate + Body Temp vs Calories")

plt.colorbar(scatter, label="Heart Rate")

plt.show()
```

```python


plt.figure(figsize=(9,7))

# Color based on error
errors = np.abs(y_test - y_pred)

scatter = plt.scatter(
    y_test,
    y_pred,
    c=errors,        # color = error
    s=80,            # dot size
    alpha=0.75
)

# Perfect prediction reference line
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    linewidth=2
)

plt.xlabel("Actual Calories", fontsize=12)
plt.ylabel("Predicted Calories", fontsize=12)
plt.title("Actual vs Predicted Calories (Colorful Comparison)", fontsize=14)

plt.colorbar(scatter, label="Prediction Error")
plt.grid(True)

plt.show()

```
