```python
import pandas as pd

data = {
    "House": [400, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1300],
    "Rent": [5000, 6500, 8000, 9500, 11000, 12500, 14000, 15500, 17000, 18500]
}

df = pd.DataFrame(data)

df
```

# modal predicted

  <!-- modal predicted -->

```python
x=df[["House"]]
y=df[["Rent"]]
x

```

```python
x_train,x_test,y_train,y_test  = train_test_split(x,y, test_size = 0.25, random_state=(42))
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()

```

```python
lr.fit(x_train,y_train)
```

```python
y_predict = lr.predict(x_test)

```

```python
plt.figure(figsize=(8,6))

# Actual full data
plt.plot(df["House"], df["Rent"], marker='o', label="Actual Data")

# Prediction (test data)
plt.plot(x_test, y_predict, marker='x', linestyle='--', label="Predicted Data")

plt.xlabel("House")
plt.ylabel("Rent")
plt.title("Model Prediction vs Rent")
plt.grid()
plt.legend()

plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2 = r2_score(y_test, y_predict)
r2
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_predict)
acc
```
